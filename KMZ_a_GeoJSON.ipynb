```python
import zipfile
import os
import geopandas as gpd
from google.colab import files
import shutil

# Crear carpeta para los GeoJSON resultantes
output_folder = 'geojson_results'
os.makedirs(output_folder, exist_ok=True)

# Paso 1: subir varios archivos KMZ (selecciona todos a la vez)
uploaded = files.upload()

for kmz_path in uploaded.keys():
    print(f"Procesando {kmz_path}...")

    # Extraer KML del KMZ
    with zipfile.ZipFile(kmz_path, 'r') as kmz:
        kml_files = [f for f in kmz.namelist() if f.endswith('.kml')]
        if not kml_files:
            print(f"No se encontró archivo KML dentro de {kmz_path}, se omite.")
            continue
        kml_filename = kml_files[0]
        extract_folder = 'temp_kml_folder'
        if os.path.exists(extract_folder):
            shutil.rmtree(extract_folder)
        kmz.extract(kml_filename, path=extract_folder)

    kml_path = os.path.join(extract_folder, kml_filename)

    # Leer KML con geopandas
    try:
        gdf = gpd.read_file(kml_path, driver='KML')
    except Exception as e:
        print(f"Error leyendo KML de {kmz_path}: {e}")
        shutil.rmtree(extract_folder)
        continue

    # Guardar como GeoJSON dentro de la carpeta output_folder
    geojson_filename = os.path.splitext(kmz_path)[0] + '.geojson'
    geojson_path = os.path.join(output_folder, geojson_filename)
    gdf.to_file(geojson_path, driver='GeoJSON')

    # Limpieza temporal
    shutil.rmtree(extract_folder)

print(f"Todos los archivos GeoJSON se guardaron en la carpeta '{output_folder}'.")

# Comprimir la carpeta con todos los GeoJSON para descargar
shutil.make_archive(output_folder, 'zip', output_folder)

# Descargar el ZIP
files.download(output_folder + '.zip')
```
